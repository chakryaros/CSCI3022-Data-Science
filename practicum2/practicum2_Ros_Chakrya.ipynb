**Part C:** Perform an **Analysis of Variance hypothesis test** in order to determine if there is evidence that there is _some_ difference among the mean ratings given to these 12 Oreo flavors. Clearly state your null and alternative hypothesis, and use the significance level identified in **Part B**. You must show **all** calculations **by hand** (and may of course use Python as a calculator, and to compute values from a distribution using the appropriate percent-point-function (ppf) or cumulative distribution function (cdf)).

In addition to showing the code for your calculations, make comments **in Markdown** explaining what you are doing.

**Solution:**

Hypothese of interest in ANOVA are as follow:

$$
\begin{array}{ll}
H_0: & \mu_1 = \mu_2 = \ldots = \mu_{k} = 0  \\
H_1: & \mu_k \neq 0 \textrm{ for at least one } k=1, \ldots, 11  \\
\end{array}
$$
Where k are the number of independent group

Using F-test:
$$ F = \dfrac{SSB/SSB_{df}}{SSW/SSW_{df}}$$

where, $SSB = \sum_{i=0}^I n_i (\bar{y}_i - \bar{\bar{y}})^2$

$SSW = \sum_{i=0}^I \sum_{j=0}^{n_i} (y_{ij} - \bar{y}_i)^2 $

**Rejection region:** 

$F \ge F_{\alpha,p,n-p-1}$ = stats.f.ppf(1-$\alpha$, dfn = I-1, dfd= N-I)

**p-value:** 1 - stats.f.cdf(F,p,n-p-1) < $\alpha$


```python
dfnew1 = df.copy()
dfnew1 = dfnew1.replace({"-":0})
dfOreo = pd.DataFrame(columns=['pb&j', 'mega stuf', "lemon", "chocolate","birthday cake","double stuf",
                               "red velvet","Dunkin' Donuts Mocha","mini","thins, coconut","mint","cinnamon bun"])

list_ = {0:'pb&j', 1:'mega stuf', 2:"lemon", 3:"chocolate",4:"birthday cake",5:"double stuf",
            6:"red velvet",7:"Dunkin' Donuts Mocha",8:"mini",9:"thins, coconut",10:"mint",11:"cinnamon bun"}

def newDataFrame(df, list_):
    df_new = pd.DataFrame(columns=['pb&j', 'mega stuf', "lemon", "chocolate","birthday cake","double stuf",
                               "red velvet","Dunkin' Donuts Mocha","mini","thins, coconut","mint","cinnamon bun"])
    
    for l in list_:
        x = df.loc[df['oreo_flavor'] == list_[l], 'rater_n':'rater_a']
        x = pd.to_numeric(x.loc[l])
        df_new[list_[l]] = x
    return df_new

dfOreo = newDataFrame(dfnew1, list_)
# dfOreo = dfOreo.dropna()
dfOreo
```

```python
f_alp = stats.f.ppf(1-0.05, 11, 54-12)
print(f_alp)
```

```python
# Get total number of points
N = 54
print(N)

# Get total number of groups
I = 12
print(I)

# Compute the grand_mean
grand_mean = np.mean(dfOreo.values) 

# Compute the between-group sum of squares 
SSB = np.sum([dfOreo[col].count()*(dfOreo[col].mean()-grand_mean)**2 for col in dfOreo.columns])

# Compute the between_group degrees of freedom 
SSB_df = I-1 

# Compute the within-group sum of squares 
SSW = np.sum([np.sum((dfOreo[col] - dfOreo[col].mean())**2) for col in dfOreo.columns])

# Compute the within_group degrees of freedom 
SSW_df = N-I 

# Compute the test statistic 
F = (SSB/SSB_df)/(SSW/SSW_df) 

# Compute the associated p-value 
pval = 1 - stats.f.cdf(F, SSB_df, SSW_df) 

print("SSB, SSB_df = {:.3f}, {}".format(SSB, SSB_df))
print("SSW, SSW_df = {:.3f}, {}".format(SSW, SSW_df))
print("F, pval = {:.5f}, {:.5f}".format(F, pval))

if F >= f_alp and pval < 0.05:
    print("Reject the null hypothesis.")
else: 
    print("Fail to reject the null hypothesis.")
```

**Rejection Rejion:**
Since $F = 1.377 < 2.025$, we fail to reject the null hypothesis. Therefore, performing at $95\%$ confidence level, we don't statistically sufficient evidence to claim that there is some difference among the mean ratings given to these 12 Oreo flavors.

**P-value:**
Since p-value $= 0.219 > 0.05$, we fail to reject the null hypothesis. Therfore, performing at $95\%$ confidence level, we don't have statistically sufficient evidence to claim that there is some difference among the mean ratings given to these 12 Oreo flavors.

**Part C**: Write a function `forward_select(df, resp_str, maxk)` that takes in the DataFrame, the name of the column corresponding to the response, and the maximum number of desired features, and returns a list of feature names corresponding to the `maxk` most important features via forward selection.  At each stage in forward selection you should add the feature whose inclusion in the model would result in the lowest sum of squared errors $(SSE)$. Use your function to determine the best $k=5$ features to include in the model. Clearly indicate which feature was added in each stage. 

**Note**: The point of this exercise is to see if you can implement **foward_select** yourself.  You may of course use canned routines like statmodels OLS, but you may not call any Python method that explicitly performs forward selection.

```python
def forward_select(df, resp_str, maxk=5):
    y = df[resp_str]
    features= list(df.columns[df.columns != resp_str])
    keep = ""
    included = []
    stage = 0
    while stage < maxk:
        min_ = 100000
        for new_column in features:
            X = df[new_column]
            X = sm.add_constant(X)
            model = sm.OLS(y, X).fit()
            sse = np.sum((y - model.predict(X))**2)
#             print("sse before: {}, {} ".format(sse, new_column))
            if sse < min_:
                min_ = sse
                keep = new_column

        included.append(keep)
        features.remove(keep)
        stage +=1
        print("stage {}, add '{}' with SSE = {:.3f}".format(stage, keep, min_))
    print("Features add:",included)
    return included
desired_features = forward_select(dfHouse, "haunted", maxk=5)
```

**Part D**: Write down the multiple linear regression model, including estimated parameters, obtained by your forward selection process. 

```python
y = dfHouse["haunted"]
Xreduce = dfHouse.loc[:, desired_features]
Xreduce = sm.add_constant(Xreduce)
reduce_model = sm.OLS(y, Xreduce).fit()
reduce_model.summary()
```

**Solution:** The multiple linear regression from the forward selection is :

$$\hat{y} = 1.1294 + (-0.095) \times \text{distance cemetery} + (-0.0013) \times \text{ice creams sold} + (0.1117) \times \text{clouds}+(-0.0451) \times \text{distance metro} + (-0.0525) \times \text{bathrooms}$$

**Part E**: Perform the appropriate statistical test at the $\alpha = 0.05$ significance level to determine whether there is a statistically significant difference between the full model with all features and the reduced model obtained by forward selection in **Part D**. 

**Solution:**

$$
\begin{array}{ll}
H_0: \beta_{age} = \beta_{area}= \beta_{cats} = \beta_{howls}  = \beta_{precipitation} = \beta_{misery index} = 0 \\
H_1: \beta_{k} \neq 0 \textrm{ for at least one } k \in \{age,area, cats, howls, precipitation, misery index\} \\
\end{array}
$$

We will calcualte a Partial F-Test of the form:

$$
F = \dfrac{(SSE_{reduce} - SSE_{full})/(p-k)}{SSE_{full}/(n-p-1)}
$$

**Rejection region:** 

$F \ge F_{\alpha,p,n-p-1}$ = stats.f.ppf(1-$\alpha$, dfn = I-1, dfd= N-I)

**p-value:** 1 - stats.f.cdf(F,p,n-p-1) < $\alpha$


```python
#full model
y_full = dfHouse["haunted"]
X_full = dfHouse.loc[:,dfHouse.columns != "haunted"]
X_full = sm.add_constant(X_full)
fullModel = sm.OLS(y_full, X_full).fit()
yhat_full = fullModel.predict(X_full)
SSE_full = np.sum((y_full - yhat_full)**2)
print("SSE full = {:.03f}".format(SSE_full))

#reduce model
y_reduce = dfHouse["haunted"]
Xreduce = dfHouse.loc[:, desired_features]
Xreduce = sm.add_constant(Xreduce)
reduceModel = sm.OLS(y_reduce, Xreduce).fit()
yhat_reduce = reduceModel.predict(Xreduce)
SSE_reduce = np.sum((y_reduce - yhat_reduce)**2)
print("SSE reduce = {:.03f}".format(SSE_reduce))
n = len(dfHouse)
p = len(dfHouse.columns) - 1
k = len(desired_features)

numerator = (SSE_reduce - SSE_full)/(p-k)
denomerator = SSE_full/(n-p-1)

F = numerator/denomerator

print("F = {:.05f}".format(F))
pvalue = 1 - stats.f.cdf(F, p-k, n-p-1)
print("p-value = {}".format(pvalue))
f_alp = stats.f.ppf(1-0.04, p-k, n-p-1)
print("F alphal = {:.05f}".format(f_alp))

if F >= f_alp and pvalue < 0.05:
    print("Reject the null hypothesis.")
else: 
    print("Fail to reject the null hypothesis.")
```

**Conclution:**

Since p-value $ < \alpha = 0.05$, we reject the null hypothesis. Therefore, we conclude that there is the sufficient evident to claim that the full model is statistically significantly better than the reduced model obtained by forward selection.
